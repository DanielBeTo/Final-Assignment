## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
%pip install matplotlib
```

```python
%pip install seaborn
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# your code goes here
QUERY = """
SELECT ConvertedComp
FROM master
"""
df1 = pd.read_sql_query(QUERY,conn)

plt.figure(figsize = (10, 6))
sns.distplot(df1['ConvertedComp'], bins = 30, kde = False)
plt.title('Distribution of Converted Comp')
plt.xlabel('Converted Comp')
plt.ylabel('Count')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = """
SELECT Age
FROM master
"""
df2 = pd.read_sql_query(QUERY,conn)

plt.figure(figsize = (10, 6))
sns.boxplot(x = df2['Age'])
plt.title('Boxplot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df3 = pd.read_sql_query(QUERY,conn)

plt.figure(figsize = (10, 6))
sns.scatterplot(x = 'Age', y = 'WorkWeekHrs', data = df3)
plt.title('Correlation between Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
df4 = pd.read_sql_query(QUERY,conn)

plt.figure(figsize = (10, 6))
sns.scatterplot(x = 'WorkWeekHrs', y = 'CodeRevHrs', size = 'Age', data = df4, hue = 'Age', sizes = (20, 250), palette = 'Set1')
plt.title('Correlation between WorkWeekHrs and CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT DatabaseDesireNextYear, COUNT(DatabaseDesireNextYear) AS Count_of_DatabaseDesireNextYear
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count_of_DatabaseDesireNextYear DESC
LIMIT 5
"""
df5 = pd.read_sql_query(QUERY,conn)

Sum = df5['Count_of_DatabaseDesireNextYear'].sum()
sizes = []

for count in df5['Count_of_DatabaseDesireNextYear']:
    sizes.append(count/Sum)

plt.figure(figsize = (10, 6))
plt.pie(sizes, labels = df5['DatabaseDesireNextYear'], autopct = '%1.2f%%', startangle = 90)
plt.title('Top 5 databases that Respondents wish to learn next year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df6 = pd.read_sql_query(QUERY,conn)
medians = df6[['WorkWeekHrs', 'CodeRevHrs']].median()
data = pd.DataFrame({'Hours':medians})

plt.figure(figsize = (10, 6))
plt.bar(data.index, data['Hours'], label = data.index, color = ['blue', 'red'])
plt.ylabel('Median Hours')
plt.title('Stacked Bar Chart of Median WorkWeekHrs and CodeRevHrs (Age 30-35)')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df7 = pd.read_sql_query(QUERY,conn)
df7 = df7.groupby('Age')['ConvertedComp'].median().reset_index()
df7

plt.figure(figsize = (10, 6))
sns.lineplot(data = df7, x = 'Age', y = 'ConvertedComp', label = 'Medians', marker = 'o')
plt.title('Median of ConvertedComp for all ages from 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = """
SELECT MainBranch, COUNT(MainBranch) AS Counts_of_MainBranch
FROM master
GROUP BY MainBranch
"""
df8 = pd.read_sql_query(QUERY,conn)
df8

plt.figure(figsize = (10, 6))
sns.barplot(data = df8, x = 'Counts_of_MainBranch', y = 'MainBranch', orient = 'h')
plt.title('Main Branches')
plt.xlabel('Counts of Main Branches')
plt.ylabel('')
plt.show()
```
